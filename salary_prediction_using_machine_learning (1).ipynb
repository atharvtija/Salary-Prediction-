```python
df=pd.read_csv('https://github.com/ybifoundation/Dataset/raw/main/Salary%20Data.csv')
```

```python
import matplotlib.pyplot as plt
x=df["Experience Years"]
y=df["Salary"]
plt.figure(figsize=(10,3))
plt.scatter(x,y)
plt.show()
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.8, random_state=2529)
```

```python
x_train.shape,x_test.shape,y_train.shape,y_test.shape
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr=LinearRegression()
lr.fit(x_train.values.reshape(-1, 1),y_train)
```

```python
lr.score(x_train.values.reshape(-1, 1),y_train)*100
```

```python
lr.predict([[5]])
```

```python
year=float(input("Enter the year of experience:"))
print(lr.predict([[year]]))
```

```python
lr.intercept_
```

```python
lr.coef_
```

```python
import matplotlib.pyplot as plt
x=df["Experience Years"]
y=df["Salary"]
plt.figure(figsize=(10,3))
plt.scatter(x,y)
plt.plot(x,lr.predict(x.values.reshape(-1, 1)),color='red')
plt.show()
```
